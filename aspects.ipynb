```python
import aspect_based_sentiment_analysis as absa
import pandas as pd
nlp = absa.load()
```

```python

aspects = ["comfort", "staff", "Food", "Entertainment", "Baggage", "Service", "Boarding", "Punctuality", "Safety", "Price", "Cleanliness", "Seat"]


def extract_top_n_aspects(tweet, n=3):
    completed_task = nlp(tweet, aspects=aspects)
    aspect_confidences = {aspect: completed_task.subtasks[aspect].scores for aspect in aspects}
    sorted_aspects = sorted(aspect_confidences, key=lambda x: aspect_confidences[x], reverse=False)
    top_n_aspects = sorted_aspects[:n]
    return top_n_aspects
```

```python
tweets = pd.read_csv('processedTweets.csv')
X = tweets['text'].astype(str)
```

```python
aspects_list = []
for i in X:
    asp = extract_top_n_aspects(i)
    print(asp)
    aspects_list.append(asp)
```

```python
import numpy as np
```

```python
tweets['aspects'] = aspects_list
```
